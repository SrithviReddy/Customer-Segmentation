```python
# ============================================================
# Customer Segmentation using K-Means Clustering
# Single Ready-to-Run Code (No external files required)
# ============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

# ------------------------------------------------------------
# 1. Generate Synthetic Customer Data
# ------------------------------------------------------------
np.random.seed(42)
n = 500

data = pd.DataFrame({
    "age": np.random.randint(18, 70, n),
    "annual_income": np.random.randint(20000, 120000, n),
    "spending_score": np.random.randint(1, 100, n)
})

print("Dataset Preview:\n")
print(data.head())

# ------------------------------------------------------------
# 2. Feature Scaling
# ------------------------------------------------------------
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# ------------------------------------------------------------
# 3. Elbow Method to Find Optimal K
# ------------------------------------------------------------
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42, n_init=10)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)

plt.figure()
plt.plot(range(1, 11), wcss)
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()

# ------------------------------------------------------------
# 4. Apply KMeans (Choose K=4)
# ------------------------------------------------------------
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
data["Cluster"] = kmeans.fit_predict(scaled_data)

print("\nCluster Distribution:\n")
print(data["Cluster"].value_counts())

# ------------------------------------------------------------
# 5. Silhouette Score
# ------------------------------------------------------------
score = silhouette_score(scaled_data, data["Cluster"])
print("\nSilhouette Score:", score)

# ------------------------------------------------------------
# 6. Visualization (Income vs Spending)
# ------------------------------------------------------------
plt.figure()
plt.scatter(data["annual_income"], data["spending_score"],
            c=data["Cluster"])
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("Customer Segments")
plt.show()

# ------------------------------------------------------------
# END
# ------------------------------------------------------------
```
